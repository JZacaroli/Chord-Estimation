```python
#Convert pitch to musical note.
from math import log2, pow
def pitch_from_freq(freq):
    #Constants
    A4 = 440
    C0 = A4*pow(2, -4.75)
    note_names = ["C", "C#", "D", "D#", "E", "F", "F#", "G", "G#", "A", "A#", "B"]
    
    #Convert to semitone value above C0.
    semitonesAboveC0 = round(12*log2(freq/C0))
    octave = semitonesAboveC0 // 12
    n = semitonesAboveC0 % 12
    return note_names[n] + str(octave)
```

```python
with open('Data.nosync/70sChopin.csv', 'r') as f:
    rdr = csv.reader(f)
    rowCount = sum(1 for row in rdr)
    print('Row count:', rowCount)
    
melodyAndTimes = []
previousMelodyNote = -1
with open('Data.nosync/70sChopin.csv', 'r') as f:
    rdr = csv.reader(f)
    for row in rdr:
        currentMelodyNote = row[1]
        if currentMelodyNote != previousMelodyNote and float(currentMelodyNote) > 0:
            melodyAndTimes += [[float(row[0]), pitch_from_freq(float(row[1]))]]

        previousMelodyNote = currentMelodyNote

for time, melody in melodyAndTimes:
    print('Time:', time, 'Melody:', melody)
```

```python
#Combine the two files, not including repeated notes/chords. Just onsets!

def findChordAtTime(time_):
    prevTime = chordsAndTimes[0][0]
    prevChord = chordsAndTimes[0][1]
    for timeAndChord in chordsAndTimes:
        if timeAndChord[0] > time_:
            return prevChord
        prevTime = timeAndChord[0]
        prevChord = timeAndChord[1]

def findMelodyAtTime(time_):
    prevTime = melodyAndTimes[0][0]
    prevMelodyNote = melodyAndTimes[0][1]
    for timeAndMelody in melodyAndTimes:
        if timeAndMelody[0] > time_:
            return prevMelodyNote
        prevTime = timeAndMelody[0]
        prevMelodyNote = timeAndMelody[1]


#Minimum resolution of 0.25s for both melody and chords
timeIncr = 1/4
#Each time step will have a chord and a melody note
prevMelodyNote= -1
prevChord = -1

#List to hold all of the outputs. [(Time_0, Chord, Melody), (Time_1, Chord, Melody), ... ]
TimesChordsAndMelody = []

for i in range(int(70*1/timeIncr)):
    timeAtThisStep = timeIncr*i
    
    chordAtThisTimeStep = findChordAtTime(timeAtThisStep)
    if chordAtThisTimeStep != prevChord:
        chordToAdd = chordAtThisTimeStep
    else:
        chordToAdd = ''
    
    melodyAtThisTimeStep = findMelodyAtTime(timeAtThisStep)
    if melodyAtThisTimeStep != prevMelodyNote:
        melodyNoteToAdd = melodyAtThisTimeStep
    else:
        melodyNoteToAdd = ''
    
    TimesChordsAndMelody += [[timeAtThisStep, chordToAdd, melodyNoteToAdd]]
    
    prevChord = chordAtThisTimeStep
    prevMelodyNote = melodyAtThisTimeStep

with open('Chopin_TimesChordsAndMelody.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['Time', 'Chord', 'Melody Note'])
    for row in TimesChordsAndMelody:
        writer.writerow(row)
```
